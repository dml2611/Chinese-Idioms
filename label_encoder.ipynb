```python
# Import the libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
# Initialise the path and load the data
path = ""
eval = pd.read_excel(path + "")
```

```python
enc_a = LabelEncoder()
enc_i = LabelEncoder()
enc_f = LabelEncoder()

Accuracy = ['inaccurate', 'poor', 'moderate', 'good', 'high']
acc = enc_a.fit(Accuracy)

Intelligibility = ['unintelligible', 'little',  'much', 'most', 'all']
intel = enc_i.fit(Intelligibility)

Fluency = ['incomprehensible', 'disfluent', 'correct', 'good', 'elegant']
flue = enc_f.fit(Fluency)
```

```python
def label_encoder(num, data):
  eval[f'A{num}'] = acc.transform(eval[f'A{num}'])
  eval[f'I{num}'] = intel.transform(eval[f'I{num}'])
  eval[f'F{num}'] = flue.transform(eval[f'F{num}'])
```

```python
for j in range(1, 10):
  exec(f'label_encoder({j}, eval)')
```
