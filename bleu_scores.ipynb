```python
# Import the libraries
import numpy as np
import nltk
import pandas as pd
from nltk.translate.bleu_score import corpus_bleu, sentence_bleu
from bleu import multi_file_bleu, multi_list_bleu
```

```python
# Initialise the data input and output paths
data_path = ""  # input file path
outfile_path = ""   # output file path
```

```python
# Create a dataframe for all model translations, in this case 9 (Google, Microsoft, DeepL, GPT4, Llama2, GLM4, VolcanoTrans, NiuTrans, Baidu)
translations  =  pd.read_excel(data_path+ "----")    # excel sheet containing reference and machine translations

# Machine translations for all models
candidate_gpt4 = translations['GPT4']
candidate_deepL = translations['Deepl']
candidate_micro = translations['Microsoft']
candidate_llama2 = translations['Llama2']
candidate_googleTrans = translations['Google']
candidate_glm4 = translations['GLM4']
candidate_niuTrans = translations['Niu']
candidate_volcTrans = translations['Volc']
candidate_baidu = translations['Baidu']

# Original translations
reference = translations['HT']
```

```python
def calc_bleu(model, name):
  score = []
  for i in range(len(reference)):
    score.append(sentence_bleu(reference[i], model[i], weights = (1,0,0,0)))
  print(name, score)
  return score
```

```python
# Calculate Bleu scores for all machine translation systems
googleTrans_bleu = calc_bleu(candidate_googleTrans, "Google Translation")
deepL_bleu = calc_bleu(candidate_deepL, "DeepL")
micro_bleu = calc_bleu(candidate_micro, "Microsoft")
volcTrans_bleu = calc_bleu(candidate_volcTrans, "Volcano Trans")
baidu_bleu = calc_bleu(candidate_baidu, "Baidu")
niutrans_bleu = calc_bleu(candidate_niuTrans, "Niu Trans")
llama2_bleu = calc_bleu(candidate_llama2, "Llama2")
glm4_bleu = calc_bleu(candidate_glm4, "GLM4")
gpt4_bleu = calc_bleu(candidate_gpt4, "GPT4")
```

```python
# Store the results in a python dictionary
dict = {'GoogleTrans': googleTrans_bleu, 'DeepL': deepL_bleu, 'Microsoft': micro_bleu, 'GPT4': gpt4_bleu, 'GLM4': glm4_bleu, 'Llama2': llama2_bleu, 'NiuTrans': niutrans_bleu, 'VolcTrans': volcTrans_bleu, 'Baidu': baidu_bleu}
```

```python
# Display the results as a python dataframe
df = pd.DataFrame(dict)
df
```

```python
df.to_excel(outfile_path + '-------', index=False)     # Enter your output file name (excel)
```
